Creating Different tytpes of Tensors ->

```python
scalar = tf.constant(10) # Creating a scalar tensor

vector = tf.constant([1, 2, 3, 4]) # Creating a vector tensor

matrix = tf.constant([[1, 2, 3],  # Creating a matrix tensor
                      [4, 5, 6]])

tensor_3d = tf.constant([
    [[1, 2, 3],            # Creating a 3-D tensor
     [4, 5, 6]],

    [[7, 8, 9],
     [10, 11, 12]]
])

print("Scalar Tensor:")
print(scalar)

print("\nVector Tensor:")
print(vector)

print("\nMatrix Tensor:")
print(matrix)

print("\n3-D Tensor:")
print(tensor_3d)
```

Performing some basic operations on the tensors ->

```python

a = tf.constant([1, 2, 3])   # Create tensors
b = tf.constant([4, 5, 6])


add = tf.add(a, b)  # Addition


sub = tf.subtract(b, a)   # Subtraction


mul = tf.multiply(a, b)   # Element-wise multiplication


div = tf.divide(b, a) # Division



m1 = tf.constant([[1, 2],
                  [3, 4]])
m2 = tf.constant([[5, 6],
                  [7, 8]])
matmul = tf.matmul(m1, m2)  # Matrix multiplication


shape = a.shape     # Shape and Rank
rank = tf.rank(a)


sum_val = tf.reduce_sum(a)  # Reduction operations
mean_val = tf.reduce_mean(a)


print("Tensor a:", a)

print("\nTensor b:", b)

print("\nAddition:", add)

print("\nSubtraction:", sub)

print("\nMultiplication:", mul)

print("\nDivision:", div)

print("\nMatrix Multiplication:\n", matmul)

print("\nShape of tensor a:", shape)

print("\nRank of tensor a:", rank)

print("\nSum of elements in tensor a:", sum_val)

print("\nMean of elements in tensor a:", mean_val)
```

**Experiment 3: Basic Tensor Operations :**

 SCALAR OPERATIONS

```python
s1 = tf.constant(4)
s2 = tf.constant(2)

print("----- SCALAR OPERATIONS -----")
print("\nAddition:", tf.add(s1, s2))
print("\nSubtraction:", tf.subtract(s1, s2))
print("\nMultiplication:", tf.multiply(s1, s2))
print("\nDivision:", tf.divide(s1, s2))
print("\nPower:", tf.pow(s1, s2))
```

VECTOR OPERATIONS

```python
v1 = tf.constant([1, 2, 3])
v2 = tf.constant([4, 5, 6])

print("\n----- VECTOR OPERATIONS -----")
print("Addition:", tf.add(v1, v2))
print("Subtraction:", tf.subtract(v2, v1))
print("Multiplication:", tf.multiply(v1, v2))
print("Division:", tf.divide(v2, v1))
print("Power:", tf.pow(v1, 2))
```

2-D MATRIX OPERATIONS

```python
m1 = tf.constant([[1, 2],
                  [3, 4]])

m2 = tf.constant([[5, 6],
                  [7, 8]])

print("\n----- 2-D MATRIX OPERATIONS -----")
print("Addition:\n", tf.add(m1, m2))
print("Subtraction:\n", tf.subtract(m2, m1))
print("Multiplication:\n", tf.multiply(m1, m2))   # element-wise
print("Division:\n", tf.divide(m2, m1))
print("Power:\n", tf.pow(m1, 2))
```

**Experiment 5: Converting NumPy Arrays to Tensors**

```python
import numpy as np

numpy_scalar = np.array(5)
tensor_scalar = tf.convert_to_tensor(numpy_scalar)

print("----- SCALAR -----")
print("\nNumPy Scalar:", numpy_scalar)
print("\nTensor Scalar:", tensor_scalar)



numpy_vector = np.array([1, 2, 3, 4])
tensor_vector = tf.convert_to_tensor(numpy_vector)


print("\nNumPy Vector:", numpy_vector)
print("\nTensor Vector:", tensor_vector)



numpy_matrix = np.array([[1, 2, 3],
                         [4, 5, 6]])
tensor_matrix = tf.convert_to_tensor(numpy_matrix)


print("\nNumPy Matrix:")
print(numpy_matrix)
print("\nTensor Matrix:")
print(tensor_matrix)
```
